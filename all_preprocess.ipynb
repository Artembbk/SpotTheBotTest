# Нижний регистр, лишние пробелы, переводы строк, не алфавитные символы

```python
text_paths = []
src_path = 'txt_texts'
dst_path = 'proccessed_texts'
for filename in os.listdir(src_path):
    shutil.copyfile(os.path.join(src_path, filename),
                    os.path.join(dst_path, filename))
```

```python
for filename in os.listdir(dst_path):
    text = ''
    print(filename)
    with open(os.path.join(dst_path, filename)) as f:
        text = f.readlines()
    text = [line.lower() for line in text]
    text = [re.sub(r'[^\w\s]',' ',line,flags=re.UNICODE) for line in text]
    text = [line.strip() for line in text]
    text = [re.sub('\d', '', line) for line in text]
    text = list(filter(None, text))
    text = list(filter(lambda line: line != '\n', text))

    with open(os.path.join(dst_path, filename), 'w') as f:
        f.write(' '.join(text))
```

```python
text_names = []
src_path = "proccessed_texts"
dst_path = 'lemmatized_texts'
for filename in os.listdir(src_path):
    shutil.copyfile(os.path.join(src_path, filename), 
                    os.path.join(dst_path, filename))
```

```python
for filename in os.listdir(dst_path):
    with open(os.path.join(dst_path, filename)) as f:
        text = f.readlines()
    
    lemmatized_text = ''
    for line in text:
        tokens = line.split()
        for i, token in enumerate(tokens):
            for part in parts:
                for ending in part['other']:
                    if token.endswith(ending):
                        if token[:-(len(ending))] + part['base'] in dic: 
                            tokens[i] = token[:-(len(ending))] + part['base']
            for pronoun in pronouns:
                for form in pronoun['other']:
                    if token == 'form':
                        tokens[i] = pronoun['base']
        lemmatized_text += ' '.join(tokens) 
                    
        
        
    with open(os.path.join(dst_path, filename), 'w') as f:
        f.write(lemmatized_text)
```

# Стоп слова

```python
counts = Counter()
directory = 'lemmatized_texts'
for filename in os.listdir(directory):
    with open(os.path.join(directory, filename)) as f:
        text = f.read()
    
    counts.update(text.split())
    
stop_words = set()
for word, num in counts.most_common(300):
    stop_words.add(word)
```

```python
src_dir = 'lemmatized_texts'
dst_dir = 'no_stop_words'
for filename in os.listdir(src_dir):
    with open(os.path.join(src_dir, filename)) as f:
        text = f.read()
    
    word_list = text.split()
    filtered = [word for word in word_list if word not in stop_words]
    
    with open(os.path.join(dst_dir, filename), 'w') as f:
        f.write(' '.join(filtered))
    
        
```
