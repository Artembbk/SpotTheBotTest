```python
import numpy
import os
import re
```

```python
with open('Maciev_A.G_Chechensko-russkiy_slovar.txt') as f:
    text = f.readlines()

words = []
for line in text:
    splitted = line.split()
    if len(splitted) < 2:
        continue
    word = splitted[0].strip()
    
        
    word = re.sub("[^а-яА-ЯIi]", "", word)
    
    if len(word) < 2:
        continue
        
    words.append(word)
        
```

```python
import wrlist

wrlist.write_list(words, 'dict.pkl')
```
