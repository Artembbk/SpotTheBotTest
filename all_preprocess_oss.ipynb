# Лемматизация

```python
dic = set(wrlist.read_list('dict.pkl'))
osnov = pd.read_csv('verbs_osnovs.csv')

with open('pronouns.txt') as f:
    pronouns = f.readlines()
    
for i, pronoun in enumerate(pronouns):
    pronouns[i] = pronoun.strip()
    
```

```python
noun_adj = {'base': '', 'other': ['ы', 'йы','æн','йæн','æй','йæ','мæ',
        'ыл','йыл','имæ','ау','йау', 'тэр', 'ты', 'тæн', 'тæм', 'тæй',
        'ты', 'тыл', 'тау', 'тимæ', 'дæр']}


verb_wish1 = {'base': '', 'other': ['ин', 'ис', 'ид', 'иккам', 'иккат', 'иккой',
            'таин','аин','таис','аис','таид','аид','таиккам','аиккам','таиккат'
             ,'аиккат','таиккой','аиккой']}

verb_wish = {'base': 'ын', 'other': ['ин', 'ис', 'ид', 'иккам', 'иккат', 'иккой',
            'таин','аин','таис','аис','таид','аид','таиккам','аиккам','таиккат'
             ,'аиккат','таиккой','аиккой']}

verb_if1 = {'base': '', 'other': [ 'он','ай','а','æм','ат','ой']}

verb_if = {'base': 'ын', 'other': [ 'он','ай','а','æм','ат','ой']}

legal_participation1 = {'base': '', 'other': ['гæ', 'гæйæ']}
legal_participation = {'base': 'ын', 'other': ['гæ', 'гæйæ']}

participle = {'base': 'ын', 'other': ['гæ', 'гæйæ']}
participle1 = {'base': '', 'other': ['æг', 'т', 'д', 'инаг']}

parts = [noun_adj, verb_wish, verb_wish1, verb_if, verb_if1, legal_participation, legal_participation1, participle, participle1]
```

```python
src_path = "proccessed_texts"
dst_path = 'lemmatized_texts'
num_cpus = 3
filenames = []
for i in range(num_cpus):
    filenames.append(list())

for i, filename in enumerate(os.listdir(src_path)):
    if not filename.startswith('.'):
        filenames[i % num_cpus].append(filename)
        shutil.copyfile(os.path.join(src_path, filename), 
                        os.path.join(dst_path, filename))
    
    
```

```python
import re

def find_substring(string, substrings):
    pattern = '(' + '|'.join(map(re.escape, substrings)) + ')'
    matches = re.findall(pattern, string)
    if matches:
        return matches
    return None


@ray.remote
def lemmatize(filenames, index):
    osnovs_list = list(osnov['present'])
    osnovs_list.extend(list(osnov['past']))
    past2present = dict(zip(list(osnov['past']), list(osnov['present'])))
    for filename in tqdm(filenames, file=open(f'tmp/lemm_oss_{index}', 'w')):
        with open(os.path.join(dst_path, filename)) as f:
            text = f.readlines()

        lemmatized_text = ''
        for line in text:
            tokens = line.split()
            for i, token in enumerate(tokens):
                flag = False

                if token in dic:
                    continue

                for pronoun in pronouns:
                    if token == pronoun:
                        tokens[i] = 'PRONOUN'
                        flag = True

                if flag:
                    continue
                
                results = find_substring(token, osnovs_list)
                if results:
                    for result in results:
                        if result in osnov['past']:
                            result = past2present[result]
                            
                        if result + 'ын' in dic:
                            tokens[i] = result + 'ын'
                            flag = True
                            

                if flag:
                    continue

                for part in parts:
                    for ending in part['other']:
                        if token.endswith(ending):
                            if token[:-(len(ending))] + part['base'] in dic: 
                                tokens[i] = token[:-(len(ending))] + part['base']
                                flag = True

            lemmatized_text += ' '.join(tokens) 


        with open(os.path.join(dst_path, filename), 'w') as f:
            f.write(lemmatized_text)
            
def lemmatize_once(filenames, index):
    osnovs_list = list(osnov['present'])
    osnovs_list.extend(list(osnov['past']))
    past2present = dict(zip(list(osnov['past']), list(osnov['present'])))
    for filename in tqdm(filenames, file=open(f'tmp/lemm_oss_{index}', 'w')):
        with open(os.path.join(dst_path, filename)) as f:
            text = f.readlines()

        lemmatized_text = ''
        for line in text:
            tokens = line.split()
            for i, token in enumerate(tokens):
                flag = False

                if token in dic:
                    continue

                for pronoun in pronouns:
                    if token == pronoun:
                        tokens[i] = 'PRONOUN'
                        flag = True

                if flag:
                    continue
                
                for part in parts:
                    for ending in part['other']:
                        if token.endswith(ending):
                            if token[:-(len(ending))] + part['base'] in dic: 
                                tokens[i] = token[:-(len(ending))] + part['base']
                                flag = True
                
                
                if flag:
                    print(token, tokens[i], 'ending')
                    continue
            
                
                results = find_substring(token, osnovs_list)
                if results:
                    for result in results:
                        if result in osnov['past']:
                            result = past2present[result]
                            
                        if result + 'ын' in dic:
                            tokens[i] = result + 'ын'
                            flag = True
                            

                if flag:
                    print(token, tokens[i], 'verb')

            
            if flag:
                print(token, tokens[i], 'ending')
                continue
            
            lemmatized_text += ' '.join(tokens) 


        with open(os.path.join(dst_path, filename), 'w') as f:
            f.write(lemmatized_text)
```

# Стоп-слова

```python
counts = Counter()
directory = 'lemmatized_texts'
for filename in os.listdir(directory):
    with open(os.path.join(directory, filename)) as f:
        text = f.read()
    
    counts.update(text.split())
    
stop_words = set()
for word, num in counts.most_common(300):
    stop_words.add(word)
```

```python
src_dir = 'lemmatized_texts'
dst_dir = 'no_stop_words'
for filename in os.listdir(src_dir):
    with open(os.path.join(src_dir, filename)) as f:
        text = f.read()
    
    word_list = text.split()
    filtered = [word for word in word_list if word not in stop_words]
    
    with open(os.path.join(dst_dir, filename), 'w') as f:
        f.write(' '.join(filtered))
    
        
```
