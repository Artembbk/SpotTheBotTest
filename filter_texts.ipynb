```python
import os

directory = 'txt_texts'
for filename in os.listdir(path=directory):
    full_filename = os.path.join(directory, filename)
    if os.path.isdir(full_filename):
        continue
    with open(full_filename) as f:
        text = f.read()
        
    if len(text.split()) < 1000:
        os.remove(full_filename)
        print(filename)
```

```python
import os
import shutil
directory = 'txt_texts'
for filename in os.listdir(path=directory):
    full_filename = os.path.join(directory, filename)
    if os.path.isdir(full_filename):
        continue
    with open(full_filename) as f:
        text = f.read()
        
    
    if text.count('æ') / len(text.split()) < 0.2 and \
       text.count('а') / len(text.split()) > 0.2 and \
       text.count('о') / len(text.split()) > 0.2:
        os.remove(full_filename)
        continue
    
    if text.count('æ') / len(text.split()) < 0.2 and \
       text.count('а') / len(text.split()) < 0.2 and \
       text.count('о') / len(text.split()) < 0.2:
        print("too few symbols", filename) 
        continue
        
    if  0.2 < text.count('æ') / len(text.split()) < 0.5:
        print('mixture', filename) 
        continue
        
    if text.count('æ') / len(text.split()) > 0.5:
        shutil.move(full_filename, os.path.join(directory, 'ok', filename))
        continue
        
```
